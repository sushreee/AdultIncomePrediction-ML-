```python
data2=data.drop(['fnlwgt'],axis=1)
ip=data2.drop(['salary'],axis=1)
op=data['salary']
```

```python
#train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(ip,op,test_size=0.1)
```

```python
#standard scalar transform
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
#Logistic regression
from sklearn.linear_model import LogisticRegression
alg1=LogisticRegression()
alg1.fit(x_train,y_train)
yp=alg1.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,recall_score,precision_score,f1_score
acc=accuracy_score(y_test,yp)
recall=recall_score(y_test,yp)
f1=f1_score(y_test,yp)
```

```python
print(acc)
print(recall)
```

```python
from sklearn.naive_bayes import GaussianNB
gb=GaussianNB()
gb.fit(x_train,y_train)
```

```python
yp1=gb.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,recall_score,precision_score,f1_score
acc1=accuracy_score(y_test,yp1)
recall1=recall_score(y_test,yp1)
f11=f1_score(y_test,yp1)
```

```python
print(acc1)
print(recall1)
print(f11)
```

```python
#knn
from sklearn.neighbors import KNeighborsClassifier
alg2=KNeighborsClassifier(n_neighbors=5)
alg2.fit(x_train,y_train)
```

```python
yp3=alg2.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,recall_score,precision_score,f1_score
acc2=accuracy_score(y_test,yp3)
recall2=recall_score(y_test,yp3)
f12=f1_score(y_test,yp3)
```

```python
print(acc2)
print(recall2)
print(f12)
```
